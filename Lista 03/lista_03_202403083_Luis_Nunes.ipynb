```python
//Questão 08

package main

import "fmt"

func DBinario(decimal int) string {
	binario := ""
	for decimal > 0 {
		resto := decimal % 2
		binario = fmt.Sprint(resto) + binario
		decimal /= 2
	}
	if binario == "" {
		return "0"
	}
	return binario
}
func main() {
	var decimal int
	for {
		_, err := fmt.Scan(&decimal)
		if err != nil {
			break
		}
		fmt.Println(DBinario(decimal))
	}
}
```

```python
//Questão 09

package main

import (
	"fmt"
	"math"
)

type Ponto struct {
	x, y, z float64
}

func distancia(p1, p2 Ponto) float64 {
	return math.Sqrt(math.Pow(p1.x-p2.x, 2) + math.Pow(p1.y-p2.y, 2) + math.Pow(p1.z-p2.z, 2))
}

func main() {
	var N int
	fmt.Scan(&N)

	pontos := make([]Ponto, N)
	for i := 0; i < N; i++ {
		fmt.Scan(&pontos[i].x, &pontos[i].y, &pontos[i].z)
	}

	for i := 1; i < N; i++ {
		dist := distancia(pontos[i-1], pontos[i])
		fmt.Printf("%.2f\n", dist)
	}
}
```

```python
//Questão 14

package main

import "fmt"

func main() {
    var N int
    fmt.Scan(&N)
    valores := make([]int, N)
    for i := 0; i < N; i++ {
        fmt.Scan(&valores[i])
    }

    Ordem(valores)
    mediana := calcularMediana(valores)
    fmt.Printf("%.2f\n", mediana)
}
func Ordem(arr []int) {
    n := len(arr)
    for i := 0; i < n-1; i++ {
        for j := 0; j < n-i-1; j++ {
            if arr[j] > arr[j+1] {
                arr[j], arr[j+1] = arr[j+1], arr[j]
            }
        }
    }
}
func calcularMediana(arr []int) float64 {
    n := len(arr)
    if n%2 == 1 {

        return float64(arr[n/2])
    } else {

        return float64(arr[n/2-1]+arr[n/2]) / 2.0
    }
}
```

```python
//Questão 15

package main

import (
	"fmt"
	"math"
)

func main() {
	var T int
	fmt.Scan(&T)

	for i := 0; i < T; i++ {
		var N int
		fmt.Scan(&N)

		vetor := make([]int, N)
		for j := 0; j < N; j++ {
			fmt.Scan(&vetor[j])
		}

		distancia, comparacoes := calcularDistanciaMinima(vetor)

		fmt.Println(distancia, comparacoes)
	}
}

func calcularDistanciaMinima(arr []int) (int, int) {
	minDistancia := math.MaxInt64
	comparacoes := 0

	for i := 0; i < len(arr)-1; i++ {
		for j := i + 1; j < len(arr); j++ {
			comparacoes++
			distancia := int(math.Abs(float64(arr[i] - arr[j])))
			if distancia < minDistancia {
				minDistancia = distancia
			}
		}
	}

	return minDistancia, comparacoes
}
```

```python
//Questão 16

package main

import 	"fmt"

func main() {
	var N, K int
	fmt.Scan(&N, &K)

	arr := make([]int, N)
	for i := 0; i < N; i++ {
		fmt.Scan(&arr[i])
	}

	if aula := cancelada(arr, K); aula {
		fmt.Println("SIM")
	} else {
		fmt.Println("NAO")
		alunos := alunosantes(arr)
		for _, idx := range alunos {
			fmt.Printf("%d \n", idx)
		}
	}
}

func cancelada(arr []int, K int) bool {
	contagem := 0
	for _, hora := range arr {
		if hora <= 0 {
			contagem++
		}
	}
	return contagem < K
}

func alunosantes(arr []int) []int {
	alunos := make([]int, 0)
	for i := len(arr) - 1; i >= 0; i-- {
		if arr[i] <= 0 {
			alunos = append(alunos, i + 1)
		}
	}
	return alunos
}
```

```python
//Questão 18

package main

import "fmt"

func main() {
    var T int
    fmt.Scan(&T)

    for i := 0; i < T; i++ {
        cpf := make([]int, 11)
        for j := 0; j < 11; j++ {
            fmt.Scan(&cpf[j])
        }
        if validarCPF(cpf) {
            fmt.Println("CPF valido")
        } else {
            fmt.Println("CPF invalido")
        }
    }
}
func validarCPF(cpf []int) bool {
    soma := 0
    for i := 0; i < 9; i++ {
        soma += cpf[i] * (i + 1)
    }
    digito1 := soma % 11
    if digito1 == 10 {
        digito1 = 0
    }
    if digito1 != cpf[9] {
        return false
    }
    soma = 0
    for i := 0; i < 9; i++ {
        soma += cpf[i] * (9 - i)
    }
    digito2 := soma % 11
    if digito2 == 10 {
        digito2 = 0
    }
    if digito2 != cpf[10] {
        return false
    }
    return true
}
```

```python
//Questão 19

package main

import (
	"fmt"
	"sort"
)

func main() {
	var N int
	fmt.Scan(&N)

	numeros := make([]int, N)
	for i := 0; i < N; i++ {
		fmt.Scan(&numeros[i])
	}
	elementosUnicos := encontrarElementosUnicos(numeros)
	sort.Ints(elementosUnicos)
	for _, elemento := range elementosUnicos {
		fmt.Println(elemento)
	}
}
func encontrarElementosUnicos(numeros []int) []int {
	elementos := make(map[int]bool)
	for _, num := range numeros {
		elementos[num] = true
	}
	elementosUnicos := make([]int, 0, len(elementos))
	for elemento := range elementos {
		elementosUnicos = append(elementosUnicos, elemento)
	}
	return elementosUnicos
}
```

```python
//Questão 20

package main

import (
	"fmt"
	"math"
)
type Ponto struct {
	X, Y, Z float64
}
func main() {
	var N int
	fmt.Scan(&N)

	pontos := make([]Ponto, N)
	for i := 0; i < N; i++ {
		fmt.Scan(&pontos[i].X, &pontos[i].Y, &pontos[i].Z)
	}

	for i := 1; i < N; i++ {
		vetor := calcularVetor(pontos[i-1], pontos[i])
		maiorValor := maiorValorAbsoluto(vetor)
		fmt.Printf("%.2f\n", maiorValor)
	}
}

func calcularVetor(a, b Ponto) Ponto {
	return Ponto{b.X - a.X, b.Y - a.Y, b.Z - a.Z}
}

func maiorValorAbsoluto(p Ponto) float64 {
	max := math.Abs(p.X)
	if math.Abs(p.Y) > max {
		max = math.Abs(p.Y)
	}
	if math.Abs(p.Z) > max {
		max = math.Abs(p.Z)
	}
	return max
}
```

```python
//Questão 22

package main

import (
	"fmt"
)

func main() {
	var n, d int

	for {
		fmt.Scan(&n, &d)
		if n == 0 && d == 0 {
			break
		}

		var numero string
		fmt.Scan(&numero)

		resultado := calcularMaiorPremio(numero, d)
		fmt.Println(resultado)
	}
}

func calcularMaiorPremio(numero string, d int) string {
	var pilha []byte

	for i := 0; i < len(numero); i++ {
		digito := numero[i]

		for len(pilha) > 0 && d > 0 && pilha[len(pilha)-1] < digito {
			pilha = pilha[:len(pilha)-1]
			d--
		}

		pilha = append(pilha, digito)
	}
	pilha = pilha[:len(pilha)-d]
	var inicio int
	for inicio < len(pilha)-1 && pilha[inicio] == '0' {
		inicio++
	}

	return string(pilha[inicio:])
}
```

```python
//Questão 24

package main

import (
	"fmt"
)

func countingSort(arr []int) {
	if len(arr) == 0 {
		return
	}

	max := arr[0]
	for _, num := range arr {
		if num > max {
			max = num
		}
	}

	count := make([]int, max+1)
	for _, num := range arr {
		count[num]++
	}
	for i := 1; i <= max; i++ {
		count[i] += count[i-1]
	}
	sorted := make([]int, len(arr))
	for i := len(arr) - 1; i >= 0; i-- {
		sorted[count[arr[i]]-1] = arr[i]
		count[arr[i]]--
	}
	copy(arr, sorted)
}

func main() {
	var n int
	for {
		fmt.Scan(&n)

		if n == 0 {
			break
		}

		arr := make([]int, n)
		for i := 0; i < n; i++ {
			fmt.Scan(&arr[i])
		}

		countingSort(arr)

		for i, num := range arr {
			fmt.Print(num)
			if i != len(arr)-1 {
				fmt.Print(" ")
			}
		}
		fmt.Println()
	}
}
```

```python
//Questão 25

package main

import (
	"fmt"
	"sort"
)

func main() {
	var sorteio [6]int
	for i := 0; i < 6; i++ {
		fmt.Scan(&sorteio[i])
	}
	sort.Ints(sorteio[:])

	var n int
	fmt.Scan(&n)

	sena := 0
	quina := 0
	quadra := 0

	for i := 0; i < n; i++ {
		var aposta [6]int
		for j := 0; j < 6; j++ {
			fmt.Scan(&aposta[j])
		}
		sort.Ints(aposta[:])

		acertos := contarAcertos(sorteio, aposta)
		switch acertos {
		case 6:
			sena++
		case 5:
			quina++
		case 4:
			quadra++
		}
	}

	imprimirResultado("sena", sena)
	imprimirResultado("quina", quina)
	imprimirResultado("quadra", quadra)
}

func contarAcertos(sorteio, aposta [6]int) int {
	acertos := 0
	i, j := 0, 0
	for i < 6 && j < 6 {
		if sorteio[i] == aposta[j] {
			acertos++
			i++
			j++
		} else if sorteio[i] < aposta[j] {
			i++
		} else {
			j++
		}
	}
	return acertos
}

func imprimirResultado(tipo string, quantidade int) {
	if quantidade == 0 {
		fmt.Printf("Nao houve acertador para %s\n", tipo)
	} else {
		fmt.Printf("Houve %d acertador(es) da %s\n", quantidade, tipo)
	}
}
```

```python
//Questão 26

package main

import (
    "fmt"
    "sort"
    )
func main() {
	var T int
	fmt.Scan(&T)

	for t := 0; t < T; t++ {
		numeros := make([]int, 9)
		for i := 0; i < 9; i++ {
			fmt.Scan(&numeros[i])
		}
		anoes := AcharAnao(numeros)
        sort.Ints(anoes)
		for _, anao := range anoes {
			fmt.Println(anao)
		}
	}
}
func AcharAnao(numeros []int) []int {
	for i := 0; i < 9; i++ {
		for j := i + 1; j < 9; j++ {
			soma := 0
			anoes := make([]int, 0, 7)
			for k := 0; k < 9; k++ {
				if k != i && k != j {
					soma += numeros[k]
					anoes = append(anoes, numeros[k])
				}
			}
			if soma == 100 {
				return anoes
			}
		}
	}
	return nil
}
```
