```python
//QUESTÃO 01
package main
import "fmt"

func main() {
	var (
		provas [8]float64
		listas [5]float64
		trabf   float64
		matricula, horas int
		somap, somal float64
	)

	for {
		fmt.Scan(&matricula)
		if matricula == -1 {
			break
		}


		somap = 0
		somal = 0

		for i := 0; i < 8; i++ {
			fmt.Scan(&provas[i])
			somap += provas[i]
		}
		for i := 0; i < 5; i++ {
			fmt.Scan(&listas[i])
			somal += listas[i]
		}
		fmt.Scan(&trabf, &horas)
		media := 0.7*(somap/8) + 0.15*(somal/5) + 0.15*trabf

		resultado := ""
		if media >= 6 && horas >= 96 {
			resultado = "APROVADO"
		} else if media >= 6 && horas < 96 {
			resultado = "REPROVADO POR FREQUENCIA"
		} else if media < 6 && horas >= 96 {
			resultado = "REPROVADO POR NOTA"
		} else if media < 6 && horas < 96 {
			resultado = "REPROVADO POR NOTA E POR FREQUENCIA"
		}
		fmt.Printf("Matricula: %d, Nota Final: %.2f, Situacao Final: %s\n", matricula, media, resultado)
	}
}
```

```python
//QUESTÃO 02
package main
import "fmt"

func main() {
    var ppa, ppb float64
    fmt.Scan(&ppa)
    fmt.Scan(&ppb)

    anos := 0
    for ppa <= ppb {
        ppa += ppa * 0.03
        ppb += ppb * 0.015
        anos++
    }

    fmt.Printf("ANOS = %d\n", anos)
}
```

```python
//QUESTÃO 04
package main
import "fmt"

func main() {
	var n, i, s float64
	var k int

	fmt.Scanln(&n)
	fmt.Scanln(&i)
	fmt.Scanln(&k)
	fmt.Scanln(&s)

	var iso, isu, im, id float64 = i, i, i, i
	fmt.Println("Tabuada de soma:")
	for j := 0; j < k; j++ {
		soma := n + iso
		fmt.Printf("%.2f + %.2f = %.2f\n", n, iso, soma)
		iso += s
	}
    fmt.Println("Tabuada de subtração:")
	for j := 0; j < k; j++ {
		sub := n - isu
		fmt.Printf("%.2f - %.2f = %.2f\n", n, isu, sub)
		isu += s
	}
	fmt.Println("Tabuada de multiplicação:")
	for j := 0; j < k; j++ {
		mul := n * im
		fmt.Printf("%.2f - %.2f = %.2f\n", n, im, mul)
		im += s
	}
		fmt.Println("Tabuada de divisão:")
	for j := 0; j < k; j++ {
		div := n/id
		fmt.Printf("%.2f - %.2f = %.2f\n", n, id, div)
		id += s
	}
}
```

```python
//QUESTÃO 05
package main

import "fmt"

func main() {
	var n int

	fmt.Scanln(&n)

	if n <= 0 {
		fmt.Println("Número de elementos inválido.")
		return
	}
	seq := make([]int, n)
	for i := 0; i < n; i++ {
		fmt.Scan(&seq[i])
	}
	ml := 0
	cl := 1

	for i := 1; i < n; i++ {
		if seq[i] > seq[i-1] {
			cl++
			if cl > ml {
				ml = cl
			}
		} else {
			cl = 1
		}
	}

	fmt.Printf("O comprimento do segmento crescente máximo é: %d\n", ml-1)
}
```

```python
//QUESTÃO 06
package main

import "fmt"

func main() {
	var n int

	fmt.Scanln(&n)

	if n <= 0 {
		fmt.Println("Número de elementos inválido.")
		return
	}

	seq := make([]int, n)
	for i := 0; i < n; i++ {
		fmt.Scan(&seq[i])
	}

	ml := 1
	cl := 1
	c := 1

	for i := 1; i < n; i++ {
		if seq[i] == seq[i-1] {
			cl++
			if cl > ml {
				ml = cl
				c = ml
			}
		} else {
			cl = 1
		}
	}

	fmt.Printf("A maior subsequência de elementos iguais possui %d elementos.\n", c)
}
```

```python
//QUESTÃO 07
package main

import "fmt"

func main() {
	var somap, somaim, numero int
    var quantidadep, quantidadeim int

	for {
		fmt.Scan(&numero)

		if numero == 0 {
			break
		}

		if numero%2 == 0 {
			somap += numero
			quantidadep++
		} else {
			somaim += numero
			quantidadeim++
		}
	}

	mediap := float64(somap) / float64(quantidadep)
	mediaim := float64(somaim) / float64(quantidadeim)

	fmt.Printf("MEDIA PAR: %.2f\n", mediap)
	fmt.Printf("MEDIA IMPAR: %.2f\n", mediaim)
}
```

```python
//QUESTÃO 08
package main

import "fmt"

func main() {
	var n int

	fmt.Scan(&n)

	if n < 2 {
		fmt.Println("Campeonato inválido!")
		return
	}

	fmt.Println("Finais possíveis:")
	k := 1
	for i := 1; i <= n; i++ {
		for j := i + 1; j <= n; j++ {
			fmt.Printf("Final %d: Time%d X Time%d\n", k, i, j)
			k++
		}
	}
}
```

```python
//QUESTÃO 09
package main

import "fmt"

func main() {
	var n int
	fmt.Scan(&n)

	if n <= 1 {
		fmt.Println("Numero invalido!")
		return
	}

	primo := true
	for i := 2; i*i <= n; i++ {
		if n%i == 0 {
			primo = false
			break
		}
	}

	if primo {
		fmt.Println("PRIMO")
	} else {
		fmt.Println("NAO PRIMO")
	}
}
```
